# Extract RAVDESS Meta

The datasets paths, genders, emotions, labels(gender_emotionm, and sentiment (polarity) are input and saved as a new dataset

```python
folder = "Dataset Meta"
ds_name = "RAVDESS"
if not os.path.isdir(os.path.join(os.getcwd(), folder)):
    os.makedirs(os.path.join(os.getcwd(), folder))


directory_list = os.listdir(RAV)
directory_list.sort()

emo = []
gender = []
path = []
actor_id = []

for i in directory_list:
  if not i == '.DS_Store':
    file_name = os.listdir(RAV + i)
    for f in file_name:
        split = f.split('.')[0].split('-')
        
        emo.append(int(split[2]))
        gender_temp = int(split[6])
        if gender_temp % 2 == 0:
            gender_temp = "female"
        else:
            gender_temp = "male"
        gender.append(gender_temp)
        path.append(RAV + i + '/' + f)
        actor_id.append(split[6])

RAV_meta = pd.DataFrame(emo, columns=['Emotion'])
RAV_meta = RAV_meta.replace(
    {1: 'neutral', 2: 'neutral', 3: 'happy', 4: 'sad', 5: 'angry', 6: 'fear', 7: 'disgust', 8: 'surprise'})
RAV_meta = pd.concat([pd.DataFrame(gender, columns=['Gender']), RAV_meta], axis=1)
RAV_meta['Labels'] = RAV_meta['Gender']+'_'+RAV_meta['Emotion']
RAV_meta['Source'] = ds_name
RAV_meta['Actor ID'] = actor_id
RAV_meta = pd.concat([RAV_meta, pd.DataFrame(path, columns=['Path'])], axis=1)
RAV_meta = RAV_meta[~RAV_meta.Labels.str.contains('surprise', na=False)]

pos = ['happy']
neg = ['fear', 'disgust', 'angry', 'sad']
neut = ['neutral']
split = RAV_meta["Labels"].str.split("_", n = 1, expand = True)
RAV_meta["Gender"] = split[0]

def applyFunc(s):
    if s == 'happy':
        return 'pos'
    elif s == 'neutral':
        return 'neut'
    else:
        return 'neg'
    return ''

RAV_meta['Polarity'] = RAV_meta['Emotion'].apply(applyFunc)
RAV_meta.to_csv(os.path.join(os.getcwd(), folder, ds_name+'.csv'), index=False)
print(f"Labels: \n\n{RAV_meta.Labels.value_counts()}\n")
print(f"Emotion: \n\n{RAV_meta.Emotion.value_counts()}\n")
print(f"Gender: \n\n{RAV_meta.Gender.value_counts()}\n")
print(f"Polarity: \n\n{RAV_meta.Polarity.value_counts()}\n")
```

# Extract CREMA-T Meta

```python
folder = "Dataset Meta"
ds_name = "CREMA"
if not os.path.isdir(os.path.join(os.getcwd(), folder)):
    os.makedirs(os.path.join(os.getcwd(), folder))

dir_list = os.listdir(CREMA)
dir_list.sort()
gender = []
emotion = []
path = []
actor_id = []
female = [1002, 1003, 1004, 1006, 1007, 1008, 1009, 1010, 1012, 1013, 1018, 1020, 1021, 1024, 1025, 1028, 1029, 1030,
          1037, 1043, 1046, 1047, 1049,
          1052, 1053, 1054, 1055, 1056, 1058, 1060, 1061, 1063, 1072, 1073, 1074, 1075, 1076, 1078, 1079, 1082, 1084,
          1089, 1091]

for i in dir_list:
  
  part = i.split('_')
  if int(part[0]) in female:
        temp = 'female'
  else:
    temp = 'male'
  gender.append(temp)
  if not i == '.DS_Store':  
    if part[2] == 'SAD' and temp == 'male':
        emotion.append(f"{temp}_sad")
    elif part[2] == 'ANG' and temp == 'male':
        emotion.append(f"{temp}_angry")
    elif part[2] == 'DIS' and temp == 'male':
        emotion.append(f"{temp}_disgust")
    elif part[2] == 'FEA' and temp == 'male':
        emotion.append(f"{temp}_fear")
    elif part[2] == 'HAP' and temp == 'male':
        emotion.append(f"{temp}_happy")
    elif part[2] == 'NEU' and temp == 'male':
        emotion.append(f"{temp}_neutral")
    elif part[2] == 'SAD' and temp == 'female':
        emotion.append(f"{temp}_sad")
    elif part[2] == 'ANG' and temp == 'female':
        emotion.append(f"{temp}_angry")
    elif part[2] == 'DIS' and temp == 'female':
        emotion.append(f"{temp}_disgust")
    elif part[2] == 'FEA' and temp == 'female':
        emotion.append(f"{temp}_fear")
    elif part[2] == 'HAP' and temp == 'female':
        emotion.append(f"{temp}_happy")
    elif part[2] == 'NEU' and temp == 'female':
        emotion.append(f"{temp}_neutral")
    else:
        emotion.append('Unknown')
    path.append(CREMA + i)
    actor_id.append(part[0])

    
CREMA_meta = pd.DataFrame(emotion, columns=['Labels'])
CREMA_meta['Source'] = ds_name
CREMA_meta['Actor ID'] = actor_id
CREMA_meta = pd.concat([CREMA_meta, pd.DataFrame(path, columns=['Path'])], axis=1)
CREMA_meta = CREMA_meta[~CREMA_meta.Labels.str.contains('surprise', na=False)]

pos = ['happy']
neg = ['fear', 'disgust', 'angry', 'sad']
neut = ['neutral']
split = CREMA_meta["Labels"].str.split("_", n = 1, expand = True)
CREMA_meta["Gender"]= split[0]
CREMA_meta["Emotion"]= split[1]
  
def applyFunc(s):
    if s == 'happy':
        return 'pos'
    elif s == 'neutral':
        return 'neut'
    else:
        return 'neg'
    return ''

CREMA_meta['Polarity'] = CREMA_meta['Emotion'].apply(applyFunc)

CREMA_meta.to_csv(os.path.join(os.getcwd(),folder,ds_name+'.csv'), index=False)

print(CREMA_meta.head)

print(f"Labels: \n\n{CREMA_meta.Labels.value_counts()}\n")
print(f"Emotion: \n\n{CREMA_meta.Emotion.value_counts()}\n")
print(f"Gender: \n\n{CREMA_meta.Gender.value_counts()}\n")
print(f"Polarity: \n\n{CREMA_meta.Polarity.value_counts()}\n")
```
